```python
import pandas as pd
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import os

# Ensure the output directory exists
output_dir = 'output_pdfs'
os.makedirs(output_dir, exist_ok=True)

def create_pdf_from_row(row):
    title = row['Title']
    summary = row.get('review/summary', '')
    review_text = row.get('review/text', '')

    # Sanitize the title to make it a valid filename
    filename = "".join([c for c in title if c.isalpha() or c.isdigit() or c==' ']).rstrip()
    filepath = os.path.join(output_dir, f"{filename}.pdf")

    c = canvas.Canvas(filepath, pagesize=letter)
    c.drawString(100, 750, f"Title: {title}")
    c.drawString(100, 730, f"Summary: {summary}")
    c.drawString(100, 710, f"Review: {review_text}")
    c.save()

# Load the CSV file
df = pd.read_csv('/Users/nguyentienhuy/Documents/SRH/CaseStudy/Data/archive/Books_rating.csv')  # Replace 'path_to_your_csv_file.csv' with your CSV file path

# Filter the DataFrame to contain only the necessary columns
df = df[['Title', 'review/summary', 'review/text']]

# Iterate over each row in the DataFrame and create a PDF
for index, row in df.iterrows():
    create_pdf_from_row(row)
```

```python
import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import SimpleDocTemplate, Paragraph

# Define the source directory containing text files and the destination directory for PDFs
source_dir = '/Users/nguyentienhuy/Documents/SRH/CaseStudy/Book_review/Data'
destination_dir = '/Users/nguyentienhuy/Documents/SRH/CaseStudy/Book_review/Data_pdf'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Get a sample style sheet
styles = getSampleStyleSheet()

# Function to convert a text file to a PDF
def text_to_pdf(text_file, pdf_file):
    doc = SimpleDocTemplate(pdf_file, pagesize=letter)
    story = []
    style = styles['Normal']

    # Open the text file and add its content to the story
    with open(text_file, 'r') as file:
        for line in file:
            # Remove leading/trailing whitespace and add a line break
            line = line.strip()
            if line:
                paragraph = Paragraph(line, style)
                story.append(paragraph)

    # Build the PDF
    doc.build(story)

# Loop through all text files in the source directory
for file_name in os.listdir(source_dir):
    if file_name.endswith('.txt'):
        base_name = os.path.splitext(file_name)[0]
        pdf_file_name = f"{base_name}.pdf"
        pdf_file_path = os.path.join(destination_dir, pdf_file_name)
        text_file_path = os.path.join(source_dir, file_name)
        
        # Convert the text file to a PDF
        text_to_pdf(text_file_path, pdf_file_path)

print("Conversion completed.")
```
